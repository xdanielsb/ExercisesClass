```python
import numpy as np
from matplotlib.pyplot import imshow
from skimage import color
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = color.rgb2gray(io.imread('lena.jpg'))
imgr = img.copy()
print(type(img))
imshow( img, cmap='gray' )
```

```python
dim = 3
kernel = np.ones((dim, dim))/(dim*dim)
print( kernel )
```

### Option 1 Complexity time O(9*n) memory O(1)

```python
rows, cols = img.shape 

for row in range(1,rows-1):
    for col in range(1,cols-1):
        img[row][col] = np.sum(kernel*imgr[row-1:row+2, col-1:col+2])
```

```python
imshow( img, cmap='gray' )
```

### Option 2 Complexity time O(n) memory O(n)

```python
img_c = img.copy()/9
```

```python
for row in range(1,rows-1):
    for col in range(1,cols-1):
        img[row][col] = np.sum(img_c[row-1:row+2, col-1:col+2])
```

```python
imshow(img, cmap='gray')
```

### Proposed kernel

```python
kernel_1 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernel_2 = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
kernel_3 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernel_4 = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
```

```python
img1, img2, img3, img4 = imgr.copy(), imgr.copy(), imgr.copy(), imgr.copy()
for row in range(1,rows-1):
    for col in range(1,cols-1):
        img1[row][col] = np.sum(kernel_1*imgr[row-1:row+2, col-1:col+2])
        img2[row][col] = np.sum(kernel_2*imgr[row-1:row+2, col-1:col+2])
        img3[row][col] = np.sum(kernel_3*imgr[row-1:row+2, col-1:col+2])
        img4[row][col] = np.sum(kernel_4*imgr[row-1:row+2, col-1:col+2])
```

```python
 plt.figure(1, figsize=(8, 6))
 plt.subplot(221)
 plt.imshow(img1,cmap='gray')
 plt.subplot(222)
 plt.imshow(img2,cmap='gray')
 plt.subplot(223)
 plt.imshow(img3,cmap='gray')
 plt.subplot(224)
 plt.imshow(img4,cmap='gray')

```
