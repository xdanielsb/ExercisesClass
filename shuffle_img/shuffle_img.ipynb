```python
__author__ = "Daniel Santos"
__date__ = "29 - 04 - 2019 "
__github__ = "xdanielsb"

import numpy as np
from matplotlib.pyplot import imshow
from skimage import color
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = color.rgb2gray(io.imread('./lena.jpg'))
imgr = img.copy()
imshow(img, cmap='gray' )
print( "Shape : " +str(img.shape))
```

```python
dim = 4
arr = [i for i in range( 0, 16)]
matrix = np.array(arr).reshape((dim,dim))
np.random.shuffle( arr)
random_matrix = np.array(arr ).reshape( (dim, dim))

print("Original Matrix")
print(matrix)

print("Random Matrix")
print( random_matrix )

img_shuffled = np.array([])
rows, cols = img.shape
partition = rows // dim 
def getCords( x):
    r1, c1 = x//4, x % 4
    return r1*partition, c1*partition

for r in range( 0, dim):
    for a in range( 0, partition):
        row = np.array([])
        for c in range (0, dim):
            r1,c1 = getCords( random_matrix[r][c])
            row = np.hstack( (row, img[r1+a, c1:c1+partition]))
        if img_shuffled.shape[0] == 0:
            img_shuffled = row
        else:
            img_shuffled = np.vstack( (img_shuffled, row))

imshow( img_shuffled , cmap='gray')
```
